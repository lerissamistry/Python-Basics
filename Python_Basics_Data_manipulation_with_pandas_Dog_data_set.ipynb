# **Data manipulation with pandas**

This notebook goes through the basics of Data Manipulation with pandas by exploring a synthetic data set on different dog breed and their characteristics.

```python
# Import pandas package
import pandas as pd
```

```python
# Upload data set from CSV file
url = "https://raw.githubusercontent.com/lerissamistry/Python-Basics/main/Dog_Breeds_Dataset.csv"

df = pd.read_csv(url)
```

```python
# View the data frame
df.head()
```

**Filtering Data:**

- Filter the dataset to find breeds with a life expectancy of more than 12 years.
- Find all large-sized dog breeds with high exercise needs.


```python
# Filter the dataset to find breeds with a life expectancy of more than 12 years.
df[df['Life Expectancy (years)'] > 12]
```

```python
# Find all large-sized dog breeds with high exercise needs.
df[(df['Size'] == 'Large') & (df['Exercise Needs'] == 'High')]
```

**Sorting Data:**

- Sort the breeds by life expectancy in descending order.
- Sort the dataset by popularity rank to see which breeds are most popular.

```python
# Sort the breeds by life expectancy in descending order.
df.sort_values(by='Life Expectancy (years)', ascending=False)
```

```python
# Sort the dataset by popularity rank to see which breeds are most popular.
df.sort_values(by='Popularity Rank', ascending=False)
```

**Grouping and Aggregating:**

- Group the breeds by size and calculate the average life expectancy for each size category.
- Find the average weight range for each size category by splitting the "Weight Range" column into numeric values.

```python
# Group the breeds by size and calculate the average life expectancy for each size category.
df.groupby('Size')['Life Expectancy (years)'].mean()
```

```python
# Find the average weight range for each size category by splitting the "Weight Range" column into numeric values.
def calculate_mean(weight_range):
    # Split the string by '-'
    min_val, max_val = map(int, weight_range.split('-'))
    # Calculate the mean
    return (min_val + max_val) / 2

df['Mean Weight (kg)'] = df['Weight Range (kg)'].apply(calculate_mean)
df.head()
```

**Adding/Modifying Columns:**

- Create a column that categorizes the shedding level as binary (e.g., Shedding Binary: High = 1, Low/Moderate = 0).


```python
# Create a column that categorizes the shedding level as binary (e.g., Shedding Binary: High = 1, Low/Moderate = 0).
import numpy as np
df['Shedding Binary'] = np.where(df['Shedding Level'] == 'High', 1, 0)
df.head()
```

```python
# Create a column that categorizes the shedding level as binary (e.g., Shedding Binary: High = 1, Low/Moderate = 0).
df['Shedding Binary Method 2'] = df["Shedding Level"].map({'High': 1, 'Moderate':0, 'Low':0})
df.head()
```

**Pivot Tables:**

- Create a pivot table that shows the average life expectancy and shedding
level for each size and country of origin.


```python
# Create a pivot table that shows the average life expectancy and shedding level for each size and country of origin.
df.pivot_table(index=['Country of Origin', 'Size'], values=['Life Expectancy (years)', 'Mean Weight (kg)'], aggfunc='mean')
```

**Renaming Columns:**

- Showcase how to rename columns to something more readable or standardized, e.g., renaming Life Expectancy (years) to Life Expectancy.

```python
# Showcase how to rename columns to something more readable or standardized, e.g., renaming Life Expectancy (years) to Life Expectancy.
df.rename(columns={'Life Expectancy (years)': 'Life Expectancy', 'Shedding Binary':'Shedding Method 1', 'Shedding Binary Method 2':'Shedding Method 2'}, inplace=True)
df.head()
```

**String Operations:**

- Extract the first trait from the Temperament column and create a new column with that value.
- Create a new column that contains the number of traits listed in the Temperament column.

```python
# Extract the first trait from the Temperament column and create a new column with that value.
def first_temperament(Temperament):
    if pd.notna(Temperament):
        return Temperament.split(',')[0].strip()
    return Temperament

df['First Temperament'] = df['Temperament'].apply(first_temperament)
df.head()
```

```python
# Create a new column that contains the number of traits listed in the Temperament column.
def number_of_temperament(Temperament):
    if pd.notna(Temperament):
        return len(Temperament.split(','))
    return Temperament

df['Number of Temperaments'] = df['Temperament'].apply(number_of_temperament)
df.head()
```

**Boolean Masking:**

- Create a mask that shows only breeds with high exercise needs and moderate shedding levels.
- Find all dog breeds where the country of origin is "Germany."

```python
# Create a mask that shows only breeds with high exercise needs and moderate shedding levels.
df[(df['Exercise Needs'] == 'High') & (df['Shedding Level'] == 'Moderate')]
```

```python
# Find all dog breeds where the country of origin is "Germany."
df[df['Country of Origin'] == 'Germany']
```

**Combining Multiple Filters:**

- Find all dog breeds that are both popular (rank < 5) and have a life expectancy greater than 12 years.

```python
# Find all dog breeds that are both popular (rank < 7) and have a life expectancy greater than 12 years.
df[(df['Popularity Rank'] < 7) & (df['Life Expectancy'] > 12)]
```

**Applying Functions:**

- Apply a custom function to extract the minimum and maximum weight from the Weight Range (kg) column and split them into two separate columns.

```python
# Apply a custom function to extract the minimum and maximum weight from the Weight Range (kg) column and split them into two separate columns.

def extract_min_max(weight_range):
    min_weight, max_weight = map(float, weight_range.split('-'))
    return min_weight, max_weight

df[['Min Weight (kg)', 'Max Weight (kg)']] = df['Weight Range (kg)'].apply(extract_min_max).apply(pd.Series)
df.head()
```

**NB.** After applying extract_min_max, each value in the "Weight Range (kg)" column has been converted into a tuple of two values, like (25, 36).

Now, the second .apply(pd.Series) step converts each tuple into a pandas Series. When you apply pd.Series to a tuple, it creates a dataframe-like structure where each element of the tuple is placed in its own column.



**Data Summary and Statistics:**

- Calculate basic statistics for numerical columns such as life expectancy and popularity rank.

```python
# Calculate basic statistics for numerical columns such as life expectancy and popularity rank.
df[['Life Expectancy', 'Popularity Rank']].describe()
```

```python
# Display summary statistics (mean, min, max) for all breeds from a specific country or size category.
df[df['Country of Origin'] == 'Germany'].describe()
```

**Handling Missing Data:**

- Add some missing values to columns like Life Expectancy or Shedding Level, and showcase how to handle missing data by:
- Filling missing values using a default value or a strategy (like the mean).
- Dropping rows/columns with missing values.

```python
df.isna().sum()
```

```python
# Randomly introduce missing values
nan_indices = df.sample(frac=0.3).index  # Randomly selecting 30% of rows
df.loc[nan_indices, 'Life Expectancy'] = np.nan
```

```python
df.isna().sum()
```

```python
# Filling missing values using a default value or a strategy (like the mean).
df['Life Expectancy'].fillna(df['Life Expectancy'].mean(), inplace=True)
df.isna().sum()
```

**Visualizations (Bonus Task):**

After manipulating the data, you can create visualizations like:

- Bar plots of life expectancy vs. breed.
- A pie chart of dog breeds by size category.
- A scatter plot of life expectancy vs. popularity rank.

```python
import matplotlib.pyplot as plt
```

```python
# Bar plots of life expectancy vs. breed.
plt.figure(figsize=(8, 6))
plt.bar(df['Breed Name'], df['Life Expectancy'])
plt.xlabel('Breed')
plt.ylabel('Life Expectancy')
plt.xticks(rotation=35)
plt.title('Life Expectancy vs. Breed')
plt.show()
```

```python
# A pie chart of dog breeds by size category.
plt.figure(figsize=(8, 6))
df['Size'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Dog Breeds by Size Category')
plt.ylabel('')
plt.show()
```

```python
# A scatter plot of life expectancy vs. popularity rank.
plt.figure(figsize=(8, 6))
plt.scatter(df['Min Weight (kg)'], df['Life Expectancy'])
plt.xlabel('Min Weight (kg)')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy vs. Min Weight (kg)')
plt.show()
```
